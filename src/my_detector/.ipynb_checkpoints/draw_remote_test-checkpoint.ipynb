## Loading label map
Label maps map indices to category names, so that when our convolution network predicts `5`, we know that this corresponds to `airplane`.  Here we use internal utility functions, but anything that returns a dictionary mapping integers to appropriate string labels would be fine

```python
category_index = label_map_util.create_category_index_from_labelmap(PATH_TO_LABELS, use_display_name=True)
```

## Helper code

```python
def load_image_into_numpy_array(image):
  (im_width, im_height) = image.size
  return np.array(image.getdata()).reshape(
      (im_height, im_width, 3)).astype(np.uint8)
```

## Draw

```python
StartIndex_of_DrawImage = 0
NLimit_of_ImageDraw = 10
NDraw = 0

print('show {} images.'.format(NLimit_of_ImageDraw))

outputs_dict = []

with open(TEST_RESULT) as fp:
    for line in fp:
        outputs_dict.append(json.loads(line))

for index, string_record in enumerate(tf.python_io.tf_record_iterator(TEST_IMAGE)):
    example = tf.train.Example()
    example.ParseFromString(string_record)

    image_bytes = io.BytesIO(example.features.feature["image/encoded"].bytes_list.value[0])
    image = Image.open(image_bytes)
    image_np = load_image_into_numpy_array(image)
    
    output_dict = outputs_dict[index]
    output_dict['detection_boxes'] = np.array(output_dict['detection_boxes'])
    output_dict['detection_classes'] = np.array(output_dict['detection_classes']).astype(np.uint8)
    output_dict['detection_scores'] = np.array(output_dict['detection_scores'])
        
    vis_util.visualize_boxes_and_labels_on_image_array(
        image_np,
        output_dict['detection_boxes'],
        output_dict['detection_classes'],
        output_dict['detection_scores'],
        category_index,
        instance_masks=output_dict.get('detection_masks'),
        use_normalized_coordinates=True,
        line_thickness=8)
  
    plt.figure(figsize=IMAGE_SIZE)
    plt.imshow(image_np)    
    
    NDraw += 1

print('{} pictures has been drawn.'.format(NDraw))
```
